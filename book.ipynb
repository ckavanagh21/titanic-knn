```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
```

```python
#import titanic dataset
data = pd.read_csv("titanic.csv")
#preprocess data 
y = data["Survived"].values.reshape(-1, 1)
X = data["Fare"].values.reshape(-1, 1)
X
```

```python
data
```

```python
#Create KNN object
model = KNeighborsClassifier(n_neighbors = 3)
#Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=94)

#Fit data
model.fit(X_train, y_train)
#Predict
model.predict(X_test)
#Evaluate
model.score(X_test, y_test)
```

```python
#Iterate over loop and find neighbor value with best accuracy
score = 0
neighbor = 0
neighbor_array = np.arange(1, 52)[0:52:2]

for i in neighbor_array:
    model = KNeighborsClassifier(n_neighbors = i)
    #Split data
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=94)

    #Fit data
    model.fit(X_train, y_train.ravel())
    #Predict
    model.predict(X_test)
    #Evaluate
    current_score = model.score(X_test, y_test)
    if current_score > score:
        score = current_score
        neighbor = i
print(neighbor)
print(score)
```
